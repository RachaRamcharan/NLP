```
import pandas as pd
df = pd.read_csv('arxiv_data.csv', nrows=1000, engine='python')
print(f"DataFrame loaded with {len(df)} rows.")
df.head()
```

```
import re

def clean_text(text):
    if not isinstance(text, str):
        return ""
    # 1. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # 2. Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # 3. Remove social media mentions (@user)
    text = re.sub(r'@\w+', '', text)
    # 4. Remove hashtags (#topic)
    text = re.sub(r'#\w+', '', text)
    # 5. Convert to lowercase
    text = text.lower()
    # 6. Remove emojis (basic regex, more comprehensive might need external library like emoji)
    emoji_pattern = re.compile(
        "[" # Start of character class
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "\U00002702-\U000027B0"
        "\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)
    # 7. Remove special characters (keep alphanumeric and spaces)
    text = re.sub(r'[^a-z0-9\s]', '', text)
    # 8. Normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_summaries'] = df['summaries'].apply(clean_text)
print("Original vs Cleaned Summaries (first 5 rows):")
print(df[['summaries', 'cleaned_summaries']].head())
```

```
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download NLTK data if not already downloaded
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')
try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('wordnet')
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def tokenize_stopwords_lemmatize(text):
    if not isinstance(text, str):
        return []
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stopwords and lemmatize
    lemmatized_tokens = [
        lemmatizer.lemmatize(word) for word in tokens if word not in stop_words
    ]
    return lemmatized_tokens

df['tokenized_lemmatized_summaries'] = df['cleaned_summaries'].apply(tokenize_stopwords_lemmatize)
print("Tokenized and Lemmatized Summaries (first 5 rows):")
print(df[['cleaned_summaries', 'tokenized_lemmatized_summaries']].head())
```

```
df['rejoined_summaries'] = df['tokenized_lemmatized_summaries'].apply(lambda x: ' '.join(x))
print("Rejoined Summaries (first 5 rows):")
print(df[['tokenized_lemmatized_summaries', 'rejoined_summaries']].head())
```

## Apply spaCy for Tokenization, Stopword Removal, and Lemmatization

### Subtask:
Define a custom Python function that utilizes the loaded spaCy model to perform tokenization, remove stopwords, and lemmatize words from the 'processed_summaries' column. This function should return a list of cleaned tokens for each summary. Store these lists in a new column called 'spacy_processed_tokens'.


**Reasoning**:
Define a custom Python function `process_text_with_spacy` that leverages the loaded spaCy model (`nlp`) to tokenize, remove stopwords and punctuation, and lemmatize words. Then, apply this function to the `processed_summaries` column to create the new `spacy_processed_tokens` column as per the instructions.



```
def process_text_with_spacy(text):
    if not isinstance(text, str):
        return []
    doc = nlp(text)
    cleaned_tokens = [
        token.lemma_ for token in doc
        if not token.is_stop and not token.is_punct and not token.is_space
    ]
    return cleaned_tokens

df['spacy_processed_tokens'] = df['processed_summaries'].apply(process_text_with_spacy)

print("Processed Summaries vs spaCy Processed Tokens (first 5 rows):")
print(df[['processed_summaries', 'spacy_processed_tokens']].head())
```
