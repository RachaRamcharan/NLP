```python
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = "Loving the new AI features 😍 #AI #MachineLearning"
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        nouns.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```

```python
import nltk
import spacy
from collections import Counter
import pandas as pd

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

nlp = spacy.load("en_core_web_sm")

text = """
Academic writing plays a crucial role in the development of critical thinking skills.
Researchers analyze data to support arguments and present findings in a structured manner.
Effective academic essays demonstrate clarity, coherence, and the appropriate use of discipline-specific vocabulary.
"""

print("SAMPLE TEXT:\n", text)

tokens = nltk.word_tokenize(text)
print("\nTOKENS:\n", tokens)

nltk_pos = nltk.pos_tag(tokens)
print("\nNLTK POS TAGGING:\n", nltk_pos)

doc = nlp(text)
spacy_pos = [(token.text, token.pos_, token.tag_) for token in doc]

print("\nspaCy POS TAGGING:\n", spacy_pos)

comparison_df = pd.DataFrame({
    "Token": [token.text for token in doc],
    "spaCy_POS": [token.pos_ for token in doc],
    "spaCy_Tag": [token.tag_ for token in doc]
})

print("\nPOS TAG COMPARISON (spaCy):")
print(comparison_df)


nltk_nouns = [word for word, pos in nltk_pos if pos.startswith('NN')]
nltk_verbs = [word for word, pos in nltk_pos if pos.startswith('VB')]

spacy_nouns = [token.text for token in doc if token.pos_ == "NOUN"]
spacy_verbs = [token.text for token in doc if token.pos_ == "VERB"]

print("\nNLTK NOUNS (Concepts):", nltk_nouns)
print("NLTK VERBS (Arguments):", nltk_verbs)

print("\nspaCy NOUNS (Concepts):", spacy_nouns)
print("spaCy VERBS (Arguments):", spacy_verbs)


noun_freq = Counter(spacy_nouns)
verb_freq = Counter(spacy_verbs)

print("\nNOUN FREQUENCY:")
for noun, freq in noun_freq.items():
    print(noun, ":", freq)

print("\nVERB FREQUENCY:")
for verb, freq in verb_freq.items():
    print(verb, ":", freq)

print("\nDISCUSSION:")
print("""
This analysis demonstrates how POS tagging can be used to study the structure and vocabulary of academic writing.
Nouns primarily represent key concepts such as 'writing', 'development', and 'vocabulary', while verbs indicate
arguments and actions such as 'analyze', 'demonstrate', and 'support'.

spaCy provides more linguistically informed tags, while NLTK offers a simpler tagset.
Such analysis helps in understanding lexical density and argumentative structure in academic texts.
""")
```
